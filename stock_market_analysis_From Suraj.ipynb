```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import  seaborn as sns
```

```python
#Preprocess the data.

#Select Relevant Features for prediction
```

```python
features =['Tesla_Vol.']
```

```python
#spilit the dataset into features (x) and target variable (y)
```

```python
x=stock_market[features]
y=stock_market['Tesla_Price']
```

```python
#Train test spilit  and linear regression model
```

```python
#Split the data into traning and testing sets

X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#initiliaze and train the linear regression model
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
#Make prediction on the test set
y_pred=model.predict(X_test)
```

```python
#Evalute the model
mse=mean_squared_error(y_test,y_pred)
print(f'mean Squared error:{mse}')
```

```python
from importlib import reload
plt=reload(plt)


#Use the trained model to predict Tesla _price on the test set
y_pred_test=model.predict(X_test)

#compare the predicated values with the actucal values
test_results=pd.DataFrame({'Actual':y_test,'Predicted':y_pred_test})
print(test_results)

#Visualize the comparsion
plt.figure(figsize=(12,6))
plt.plot(test_results['Actual'],label='Actual Tesla_Price',marker='o')
plt.plot(test_results['Predicted'],label='Predicted Tesla_Price',marker='x')
plt.title('Comparision of Actual vs Predicted Tesla_Price')
plt.xlabel('Sample Index')
plt.ylabel('Tesla_Price(USD)')

plt.show()
```
